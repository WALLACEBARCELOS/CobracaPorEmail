```python
import win32com.client as client
import pandas as pd
import datetime as dt
```

```python
#le a atabela e manda os dados para 'tabela'
tabela = pd.read_excel("Contas a Receber.xlsx")
#exibe a tabela
display(tabela)
#apresenta a iformação de valores da tabela
tabela.info()
```

```python
#armazena a data de hoje
hoje = dt.datetime.now()
print (hoje)
```

```python
#pega os devedores com o status em andamento
tabela_devedores = tabela.loc[tabela['Status']=='Em aberto']
#apresenta
display(tabela_devedores)
#Pega os devedores que estao com a data vencida de hoje
tabela_devedores = tabela_devedores.loc[tabela_devedores['Data Prevista para pagamento']<hoje]
display(tabela_devedores)
```

```python
#Abre a aplicação outlook
outlook = client.Dispatch ('Outlook.Application')
```

```python
#Atribui o email do remetente
emissor = outlook.session.Accounts['wallacebarcelos@ocyan-partner.com']
```

```python
#cria um novo email
mensagem = outlook.CreateItem(0)
#mostra na tela
mensagem.Display()
#destinatario
mensagem.To = 'wallacebarcelos5@gmail.com'
#assunto
mensagem.Subject = 'Atraso no Pagamento'
#Corpo do email
mensagem.Body = '''

VOU ESCREVER O QUE EU QUISER AQUI

'''
#selecionar o emisor para enviar email
mensagem._oleobj_.Invoke(*(64209,0,8,0,emissor))
#salva
mensagem.Save()
#envia
mensagem.Send()
```

```python
dados = tabela_devedores[['Valor em aberto','Data Prevista para pagamento', 'E-mail', 'NF']].values 
```

```python
for dado in dados:
    
    destinatario = dado[2]
    nf=dado[3]
    prazo=dado[1]
    prazo = prazo.strftime("%d/%m/%Y")
    valor=dado[0]
    
    #cria um novo email
    mensagem = outlook.CreateItem(0)
    #mostra na tela
    mensagem.Display()
    #destinatario
    mensagem.To = destinatario
    #assunto
    mensagem.Subject = 'Atraso no Pagamento'
    #Corpo do email
    mensagem.Body = f'''

        Prezado Cliente,

        Verificamos um atraso no pagamento referente a NF {nf} com vencimento em {prazo} e valor total de R${valor:.2f}.
        Gostaríamos de verificar se há algum problema que necessite de auxílio de nossa equipe. 

        Em caso de dúvidas, é só entrar em contato com nosso time atráves do e-mail tecobreiautomaticamentecompython@gmail.com


        Att,
        Wallace barcelos 
    '''
        
    #selecionar o emisor para enviar email
    mensagem._oleobj_.Invoke(*(64209,0,8,0,emissor))
    #salva
    mensagem.Save()
    #envia
    mensagem.Send()
```
